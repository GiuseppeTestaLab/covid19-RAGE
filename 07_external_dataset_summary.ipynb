```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set_theme(font_scale=1.5)
sns.set_style("whitegrid")
```

```python
results = pd.read_csv("summary_results.csv")
results
```

```python
results["name"] = results["cfr"] + " - " + results["dataset"]
results["log10_pval"] = -np.log10(results["pvalue"])
```

```python
ax = sns.relplot(data=results.sort_values(by="name"), x="NES", y="name", size="log10_pval", 
                 sizes=(10,500), size_norm=(0.,3.2), 
                 hue="log10_pval", hue_norm=(0.2,3.2), palette="copper", 
                 height=9, aspect=1.2, legend="auto")
```
